```python
import matplotlib as mpl
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import netCDF4
from IPython.display import display, Math, Latex
from numpy.random import rand
from pylab import rcParams, pcolor, show,colorbar, xticks, yticks, title, figure
from IPython.display import Image
from sklearn.neighbors import KernelDensity
from sklearn.model_selection import GridSearchCV
import csv
```

```python
file = r'C:\\Users\\rhamm\\Dropbox\\RSMAS\\AppliedDataAnalysis\\hurricanes_data.csv'
data = np.genfromtxt(file,delimiter=',',dtype=None, encoding=None,names=True)
major = data['major_hurricanes']
ace = data['ace']
area = data['sp_area']

```

```python
# Define my scatter plot and container
mpl.rcParams.update(mpl.rcParamsDefault)
%matplotlib inline
rcParams['figure.figsize'] = (11, 6.5)
def fig1():
    fig,ax=plt.subplots(1,1)
    ax.plot(area,ace,'.')
    ax.set_xlabel('Spring Sea Ice Area'); ax.set_ylabel('ACE')
    ax.set_title('Spring Sea Ice and ACE')
def fig2():
    fig,ax=plt.subplots(1,1)
    ax.plot(area,major,'.')
    ax.set_xlabel('Spring Sea Ice Area'); ax.set_ylabel('# Major Hurricanes (cat 3, 4, or 5)')
    ax.set_title('Spring Sea Ice and Major Hurricanes')
```

```python
fig1()
fig2()

```

```python
import matplotlib.cm as cm  # import the colormaps.\n",
fig,ax=plt.subplots(1,1)    # Create a result-shaped container (a fig)

# Make the 2D histogram with 100 bins
x = area
y = ace
(n,xx,yy,im)=ax.hist2d(x,y,bins=100,cmap=cm.Reds,cmin=0.,normed=True)
   
# Plot it, with a colorbar
plt.colorbar(im,shrink=0.5)
ax.set_xlabel('Spring Sea Ice Area'); ax.set_ylabel('ACE')
ax.set_title('Spring Sea Ice and ACE')
plt.show()
```

```python
# Do it in Seaborn, with the marginal distributions on the top and side
import seaborn as sns
sns.set(color_codes=True)

df = pd.DataFrame({'x': x, 'y': y})
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
```
